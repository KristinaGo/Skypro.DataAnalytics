```python
##############################################   Задание 2.1   #################################################################

##### Импортните csv-файл под названием Purchase_List
##### Приведите date_purchase к формату даты с помощью функции to_datetime
##### Создайте колонку датафрейма с месяцем оплат (числом, от 1 до 12)
##### Сделайте группировку: количество оплат и сумма оплат для каждого месяца (создайте новый группировочный датафрейм)
##### Постройте визуализацию: для каждого месяца покупки по левой оси постройте сумму оплат, а по правой - количество оплат
##### Не забудьте оформление графика (заглавие, подпись осей, раздельная легенда, сетка)

import pandas as pd
import datetime as dt

purchase_list = pd.read_csv('Purchase_List.csv')
purchase_list['date_purchase'] = pd.to_datetime(purchase_list['date_purchase']) #привели колонку даты к типу дата
#purchase_list.info()

purchase_list['month'] = purchase_list['date_purchase'].dt.month #создали новую колонку с номером месяца
purchase_per_month = purchase_list.groupby(['month']).id_purchase.count().reset_index()
purchase_amt = purchase_list.groupby(['month']).amt_pay.sum().reset_index()

#purchase_list
#purchase_per_month
new_purchase = pd.merge(purchase_per_month, purchase_amt, how = 'right',  on = 'month')

#new_purchase.set_index('month', inplace = True) 

new_purchase.reset_index()
new_purchase
```

```python
mnth = list(new_purchase['month'])
pay = list(new_purchase['amt_pay'])
cnt = list(new_purchase['id_purchase'])
```

```python
fig, ax1 = plt.subplots()

ax1.set_title('график')
ax1.set_xlabel('Месяц')
ax1.set_ylabel('Сумма оплат')
ax1.plot(mnth, pay, 'm', label = 'Сумма оплат')
ax1.legend(loc = 6)

ax1.grid()
ax2 = ax1.twinx()

ax2.set_ylabel('Количество оплат')
ax2.plot(mnth, cnt, 'g', label = 'Количество оплат')
ax2.legend(loc = 5)

plt.show()
```
