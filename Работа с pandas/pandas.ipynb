```python
##############################################   Задание 1.3   #################################################################

########  1. Постройте группировку: для каждого города найдите количество клиентов. 
################   Сохраните результат в датафрейм client_pool_city


client_pool_city = client_pool.groupby('id_city')['id_client'].count()

client_pool_city
########  2. Постройте группировку: для каждого пола найдите количество клиентов
################   Сохраните результат в датафрейм client_pool_gender

########  3. Постройте группировку: найдите количество клиентов для каждой комбинации пол-город
################   Сохраните результат в датафрейм client_pool_city_gender

```

```python
client_pool_gender = client_pool.groupby('gender')['id_client'].count()

client_pool_gender
```

```python
client_pool_city_gender = client_pool.groupby( ['gender', 'id_city'])['id_client'].count().reset_index()

client_pool_city_gender
```

```python
#3. Постройте группировку: найдите количество клиентов для каждой комбинации пол-город
################   Сохраните результат в датафрейм client_pool_city_gender

client_pool_city_gender1 = pd.merge(client_pool_city_gender, city_dict)
client_pool_city_gender1
```

```python
##############################################   Задание 1.4   #################################################################

#######  С помощью функции merge присоедините таблицу city_dict к таблице client_pool_city, чтобы в ней была расшифровка городов
#######  Результат запишите в датафрейм client_pool_city_f
#######  В результирующем датафрейме client_pool_city_f не должно быть колонки id_city

client_pool_city_f = pd.merge(client_pool_city, city_dict, 
                              how = 'left', left_on='id_city', right_on='id_city').drop(columns = 'id_city')

client_pool_city_f
```

```python
##############################################   Задание 1.5   #################################################################

#######  С помощью функции merge присоедините таблицу city_dict к таблице client_pool_city_gender, чтобы в ней была расшифровка городов
#######  Результат запишите в датафрейм client_pool_city_gender_f. 
#######  В результирующем датафрейме client_pool_city_gender_f не должно быть колонки id_city

client_pool_city_gender_f = client_pool_city_gender1.drop(columns = 'id_city')
client_pool_city_gender_f
```

```python
##############################################   Задание 1.6   #################################################################

#######  Создайте новый датафрейм client_pool_city_gender_f_women, в который запишите 
#######  строчки датафрейма client_pool_city_gender_f, соответствующие женскому полу (с помощью слайсования датафрейма)
#######  Переименуйте поле id_client в client_pool_city_gender_f_women на "cnt_women" 

#######  Hint: Можно освежить метод rename по ссылке: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rename.html
client_pool_city_gender_f_women = client_pool_city_gender_f.loc[client_pool_city_gender_f['gender'] == 'F']

client_pool_city_gender_f_women = client_pool_city_gender_f_women.rename(columns={'id_client': 'cnt_women'})

client_pool_city_gender_f_women
```

```python
##############################################   Задание 1.7   #################################################################

#######  Создайте новый датафрейм client_pool_city_gender_f_men, в который запишите 
#######  строчки датафрейма client_pool_city_gender_f, соответствующие мужскому полу (с помощью слайсования датафрейма)
#######  Переименуйте поле id_client в client_pool_city_gender_f_men на "cnt_men" 

#######  Hint: Можно освежить метод rename по ссылке: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rename.html
client_pool_city_gender_f_men = client_pool_city_gender_f.loc[client_pool_city_gender_f['gender'] == 'M']

client_pool_city_gender_f_men = client_pool_city_gender_f_men.rename(columns={'id_client': 'cnt_men'})

client_pool_city_gender_f_men
```

```python
##############################################   Задание 1.8   #################################################################

#######   Соедините датафреймы client_pool_city_gender_f_men и client_pool_city_gender_f_women с помощью метода merge.
#######   Результат запишите в новый датафрейм client_pool_city_gender_gencol
#######   Добавьте в результирующий датафрейм client_pool_city_gender_gencol колонку cnt_all, которая покажет суммарное количество людей в городе
#######   В результирующем датафрейме client_pool_city_gender_gencol должны быть колонки ['name_city','cnt_men','cnt_women', 'cnt_all']
client_pool_city_gender_gencol = pd.merge(client_pool_city_gender_f_women, client_pool_city_gender_f_men,
                                          how = 'left', left_on = 'name_city', right_on = 'name_city')
#client_pool_city_gender_gencol = client_pool_city_gender_gencol.groupby('name_city')[['cnt_women','cnt_men']].sum()

client_pool_city_gender_gencol['cnt_all'] = client_pool_city_gender_gencol['cnt_women'] + client_pool_city_gender_gencol['cnt_men']
client_pool_city_gender_gencol = client_pool_city_gender_gencol.groupby('name_city')[['cnt_women','cnt_men', 'cnt_all']].sum()

client_pool_city_gender_gencol

sorted_client_pool_city_gender_gencol = client_pool_city_gender_gencol.sort_values(by ='cnt_all', ascending = False)
sorted_client_pool_city_gender_gencol
```
