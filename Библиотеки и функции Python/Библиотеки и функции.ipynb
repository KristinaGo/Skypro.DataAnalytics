### Задание 1. Функция для списка

```python
##############################################   Задание 1.1   #################################################################


##### Создайте функцию function_for_list, которая на вход будет принимать список и с помощью функции print будет выдавать:
#######  1. Сумму элементов списка
#######  2. Максимальный элемент списка
#######  3. Минимальный элемент списка
#######  4. Длину списка
#######  5. Среднее значение списка

#####  Убедитесь, что все рассчитанные значения сохраняются во внутренние переменные (внутри функции) для дальнейших манипуляций
#####  Прогоните вашу функцию на списке lst = [1, 2, 3]


def function_for_list(lst): 
  
    sum_lst = sum(lst)
    max_lst = max(lst)
    min_lst = min(lst)
    len_lst = len(lst)
    mean_lst = mean(lst)
    

    print('Сумму элементов списка :' , sum_lst)
    print('Максимальный элемент списка :' , max_lst)
    print('Минимальный элемент списка :' , min_lst)
    print('Длина списка :' , len_lst)
    print('Среднее значение списка :' , mean_lst)
    
      
```

### Задание 2. Функция для словаря

```python
##############################################   Задание 2.1   #################################################################


##### Создайте функцию function_for_dict, которая на вход будет принимать словарь и с помощью функции print будет выдавать:
#######  1. Количество элементов словаря
#######  2. Типы данных ключей словаря
#######  3. Сумму значений словаря (с помощью цикла, пробегающего по всем элементам словаря)

#####  Убедитесь, что рассчитанное значение сохраняется во внутреннюю переменную (внутри функции) для дальнейших манипуляций
#####  Прогоните вашу функцию на словарях: 
#dct = {'Ivan':50000, 'Mary':75000, 'Michael': 64000, 'Anna':80000}
#dct = {14:50000, 'Mary':75000, 64.99: 64000}

def function_for_dict(dct):
    count_lendict = len(dct)
    print('Длина словаря', count_lendict)

    sum_values = 0
    for i in dct:
        print(type(i))
        sum_values += dct[i]  
    print('Сумма значений -', sum_values )
    
```

```python
dct = {'Ivan':50000, 'Mary':75000, 'Michael': 64000, 'Anna':80000}
function_for_dict(dct)
```

```python
dct = {14:50000, 'Mary':75000, 64.99: 64000}
function_for_dict(dct)
```

```python
##############################################   Задание 2.2   #################################################################

#### Создайте внутри функции (в начале) счетчик, который подсчитывает количество не строчных ключей словаря
#### Добавьте в вашу функцию условие: если значение этого счетчика больше 0, то функция принтит "Ключами должны быть имена!"
#### В противном случае функция воспроизводит действия из пункта 2.1 (назовем их "основной командой")

#####  Прогоните вашу функцию на словарях: 
##dct = {'Ivan':45, 'Mary':38, 'Michael': 64, 'Anna':22}
#######   2. dct = {15.65:45, 'Mary':38, 'Michael': 64, 'Anna':22}


def function_for_dict(dct):
    counter = 0
    for i in dct:
        if type(i) != str:
            counter += 1
    if counter > 0:
        print('Ключами должны быть имена!')
    else:
        count_lendict = len(dct)
        print('Длина словаря', count_lendict)

        sum_values = 0
        for i in dct:
            print(type(i))
            sum_values += dct[i]  
        print('Сумма значений -', sum_values )
    
```

```python
dct = {15.65:45, 'Mary':38, 'Michael': 64, 'Anna':22}
function_for_dict(dct)
```

```python
##############################################   Задание 2.3   #################################################################

#### Создайте внутри функции счетчик, который подсчитывает количество не цифровых значений словаря
#### Добавьте в вашу функцию условие: если есть не цифровые значения словаря, то функция принтит "Нужны числа в значениях!"
#### В противном случае функция воспроизводит "основную команду".Нужны числа в значениях!
#### Код, добавленный в пункте 2.2, остается в самом начале функции

##### Подсказка: условие на тип данных значений будет внутренним условием относительно проверки на тип данных ключей словаря.

#####  Прогоните вашу функцию на словарях: 
##dct = {'Ivan':45, 'Mary':38, 'Michael': 64, 'Anna':22}
#######   2. dct = {15.65:45, 'Mary':38, 'Michael': 64, 'Anna':22}
#######   3. dct = {'Ivan':45, 'Mary':'Mary', 'Michael': 64, 'Anna':22}
#######   4. dct = {15.65:45, 'Mary':38, 'Michael': 64, 'Anna':'22'}


def function_for_dict(dct):
   
    counter_keys = 0
    counter_values = 0
    
    for i in dct:
        if type(i) != str:
            counter_keys += 1
        if (type(dct[i]) != int) & (type(dct[i]) != float):
            counter_values += 1
            
    if counter_keys > 0:
        print('Ключами должны быть имена!')
    else:
        if counter_values > 0:
            print('Нужны числа в значениях')
        else:

            count_lendict = len(dct)
            print('Длина словаря', count_lendict)

            sum_values = 0
            for i in dct:
                print(type(i))
                sum_values += dct[i]  
            print('Сумма значений -', sum_values )
```

```python
dct = {'Ivan':45, 'Mary':38, 'Michael': 64, 'Anna':22}
function_for_dict(dct)
```

### Задание 3. Функция для алерта по метрикам

```python
##############################################   Задание 3.1   #################################################################


##### Создайте функцию metric_alert, которая на вход будет принимать словарь с датами и значениями конверсии и будет:
#########    1. Определять среднее арифметическое значение метрики для всех дат в словаре (с помощью цикла) и принтить его
#########    2. Находить те значения метрик, которые больше чем на 10 процентных пунктов отличаются от общего среднего
#########    3. Принтить даты, соответствующие сильно отклоняющимся значениям метрики

######### Подсказка: Нужны значения метрик, которые отклоняются как вверх, так и вниз. 
#########            Можно использовать функцию abs (абсолютное значение, модуль)
#####  Прогоните вашу функцию на словаре: 
#dict_metric = {'01.01.2021':0.46 , '01.02.2021':0.64, '01.03.2021':0.5, '01.04.2021':0.61 , 
#               '01.05.2021':0.66 , '01.06.2021':0.4, '01.07.2021':0.32, '01.08.2021':0.66 ,
#               '01.09.2021':0.68 , '01.10.2021':0.54, '01.11.2021':0.57, '01.12.2021':0.59 }


def metric_alert(dict_metric):
    sum_metric = 0
    
    for i in dict_metric:
        sum_metric += dict_metric[i]
        
    avg_metric = sum_metric / len(dict_metric)
    print('Среднее значение конверсии - ', avg_metric)
    print()
    
    list_alerts = []
    for i in dict_metric:
        if abs(dict_metric[i] - avg_metric) >= 0.1:
            list_alerts.append(i)
            
    print('Месяцы с отклонениями:', list_alerts)
```

```python
dict_metric = {'01.01.2021':0.46 , '01.02.2021':0.64, '01.03.2021':0.5, '01.04.2021':0.61 , 
               '01.05.2021':0.66 , '01.06.2021':0.4, '01.07.2021':0.32, '01.08.2021':0.66 ,
               '01.09.2021':0.68 , '01.10.2021':0.54, '01.11.2021':0.57, '01.12.2021':0.59 }

metric_alert(dict_metric)
```
